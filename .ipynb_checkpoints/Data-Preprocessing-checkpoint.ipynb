# READ DATA COLLECTED FROM GOOGLE MAPS

```python
df = pd.DataFrame(columns = ["day","time","longtitude_start","latitude_start","longtitude_end","latitude_end","details"])
segment_i = "0"
Months = ["January", "February", "March", "April", "May", "June", "July", "August", "sebtember", "October", "November", "Decemeber"]
for month in Months:
    longtitude_start,latitude_start = "-71.293671", "46.764391"
    longtitude_end, latitude_end = "-71.288676", "46.76698"
    url1 = "Robert-bourassa-ULaval/segment:"+month+","+segment_i+"_"+longtitude_start+","+latitude_start+"-"+longtitude_end+","+latitude_end+".txt"
    with open(url1, "r") as file:
        data = json.load(file)
    df_index = pd.DataFrame(data)
    df_index["longtitude_start"] = [longtitude_start for i in range(len(df_index))] 
    df_index["latitude_start"] = [latitude_start for i in range(len(df_index))] 
    df_index["longtitude_end"] = [longtitude_end for i in range(len(df_index))] 
    df_index["latitude_end"] = [latitude_end for i in range(len(df_index))] 
    df_index["segment"] = ["segment_"+str(segment_i) for i in range(len(df_index))] 

    #df.details.unique()
    df = pd.concat([df, df_index])
    df.index = range(len(df))
df = df.drop_duplicates()
df
```

```python
def mean_of_range(range_string):
    start, end = map(int, range_string.split(","))
    numbers = list(range(start, end+1))
    return statistics.mean(numbers)
def replace_letter(string, index, replacement):
    string_list = list(string)
    string_list[index] = replacement
    return "".join(string_list)
```

```python
# create new columns in the dataframe to store the extracted values
df['travel_time'] = None
df['distance'] = None
df['route'] = None
# iterate through each row in the "details" column
for i in range(len(df)):
#    details = df.iloc[i,2]
    #print(details)
    details = df.iloc[i,6]

    # Find the travel time
    if(re.search(r'typically (\d+–\d+) min', details)):
        #map(int, range_string.split("-"))
        travel_time = re.search(r'typically (\d+–\d+) min', details).group(1)
        #print(len(travel_time))
        #print(type(travel_time))
        if(len(travel_time) == 4):
            travel_time = replace_letter(travel_time, 1, ',')
        else:
            travel_time = replace_letter(travel_time, 2, ',')
        travel_time = mean_of_range(travel_time)
        #print(travel_time)
        #travel_time = mean_of_range(re.search(r'typically (\d+–\d+) min', details).group(1))
        #print("True 1")
    elif re.search(r'typically (\d+ min)\s', details):
        travel_time = re.search(r'typically (\d+ min)\s', details).group(1)
        travel_time = float(travel_time.split(" ")[0])
        #print("True 2")
    else:
        travel_time = None
    # Find the distance
    if(re.search(r'(\d+\.\d+) km', details)):
        distance = re.search(r'(\d+\.\d+) km', details).group(1)
    else: 
        distance = None
    
    # Find the route
    if(re.search(r'via (.+?)Details', details)):
        route = re.search(r'via (.+?)Details', details).group(1).strip()
    else:
        route = None
    
    df.at[i, 'travel_time'] = travel_time
    df.at[i, 'distance'] = distance
    df.at[i, 'route'] = route
    #print(i, travel_time,distance,route )
# show the updated dataframe
#print(df)
df
```

```python
df = df.drop_duplicates()

def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

df["season"] = df["month"].apply(get_season)

df['travel_time'].fillna(method='ffill', inplace=True)
df['distance'] = df['distance'].astype(float)
# Add a new column for speed (distance / travel_time)
df['speed'] = df['distance'] / df['travel_time']*100
#new_columns = ['day', 'month','year','hour','season','route' ,'distance','speed', 'travel_time']
df.to_csv("traffic_travel_time.csv")
```

# **Months scrapping**

```python
def month_to_number(month_name):
    month_list = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    return int(month_list.index(month_name) + 1)
```

```python
months = ['January','February', 'March', 'April','May', 'June', 'July','August', 'September', 'October','November', 'December']
```

```python
from datetime import datetime
rows = []
for month in months:
    with open("Weather_Data/"+month+".txt", "r") as file:
        data = file.read()
    day_number = 1
    for day in data.split("&"):
        for time in day.split("\n"):
            month_number = month_to_number(month)
            specific_datetime = datetime(2022,month_number, day_number)
            row = [specific_datetime] + time.split("\t")
            #print(row)
            rows = rows + [row]
        day_number = day_number+1
```

```python
df = pd.DataFrame(rows, columns=['Day','Time', 'Temperature', 'Dew Point', 'Humidity', 'Wind', 'Wind Speed', 'Wind Gust', 'Pressure', 'Precip', 'Condition'])
df = df.dropna()
```

```python
df['Day'] = pd.to_datetime(df['Day'])
```

```python
df['Temperature'] = df['Temperature'].str.replace(' °F', '')
df['Dew Point'] = df['Dew Point'].str.replace(' °F', '')
df['Wind Speed'] = df['Wind Speed'].str.replace(' mph', '')
df['Humidity'] = df['Humidity'].str.replace(' %', '')
df['Temperature'] = df['Temperature'].str.replace(' �F', '')
df['Dew Point'] = df['Dew Point'].str.replace(' �F', '')
df['Wind Gust'] = df['Wind Speed'].str.replace(' mph', '')
df['Pressure'] = df['Pressure'].str.replace(' in', '')
df['Precip'] = df['Precip'].str.replace(' in', '')
```

```python
df.replace(' N/A', np.nan, inplace=True)
```

```python
df['Temperature'] = df['Temperature'].astype(float)
df['Dew Point'] = df['Dew Point'].astype(float)
df['Humidity'] = df['Humidity'].astype(int)
df['Wind Speed'] = df['Wind Speed'].astype(float)
df['Wind Gust'] = df['Wind Gust'].astype(float)
df['Pressure'] = df['Pressure'].astype(float)
df['Precip'] = df['Precip'].astype(float)
df = df.dropna()
```

```python
df['Time'][0].split(' ')
def AM__PM__24_Hours(time):
    if((time.split(' ')[1] == 'AM')&(time.split(' ')[0].split(':')[0] == "12")):
        return 0
    if(time.split(' ')[1] == 'AM'):
        return time.split(' ')[0].split(":")[0]
    else:
        return int(time.split(' ')[0].split(":")[0]) + 12
df["Time"] = df["Time"].apply(AM__PM__24_Hours)
df.rename(columns={'Time': 'hour'},inplace=True, errors='raise')
df
```
